```python
import pandas as pd
```

```python
from hmcollab import datasets
from hmcollab import articles
from hmcollab import transactions
```

```python
dataset = datasets.HMDataset()
```

```python
customer_1 = dataset.transactions.customer_id[0]
```

```python
features = [
        "product_type_no",
        "product_group_name",
        "graphical_appearance_no",
        "colour_group_code",
        "perceived_colour_value_id",
        "perceived_colour_master_id",
        "department_no",
        "index_code",
        "index_group_no",
        "section_no",
        "garment_group_no",
        ]
```

```python
def dummy_features(df, columns):
    # for articles
    return pd.get_dummies(df[['article_id'] + columns], columns=columns, prefix=columns)
    
```

```python
dummies = dummy_features(dataset.articles, features)
```

```python
average_customer_1 = transactions.kmeans_consumer(customer_1, dataset.transactions, dummies)
print('Customer:', customer_1)
print('Dimensionality of vector:', len(average_customer_1.cluster_centers_[0]))
print('kmeans vector for k=1')
average_customer_1.cluster_centers_[0]
```
