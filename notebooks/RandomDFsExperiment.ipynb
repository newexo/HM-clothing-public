```python
from hmcollab import directories
from hmcollab.datasets import HMDataset
from hmcollab.directory_tree import HMDatasetDirectoryTree
from hmcollab import splitter

import numpy as np
import pandas as pd
```

```python
test_data_tree = HMDatasetDirectoryTree(
    base=directories.testdata("fivehundred")
)
test_data = HMDataset(tree=test_data_tree)
```

```python
selected_customers = test_data.customers.customer_id.unique()[:10]
portion = splitter.CustomerPortion(selected_customers)
trimmed_data = portion.split(test_data)
```
