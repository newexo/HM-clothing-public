```python
import pandas as pd
import numpy as np
```

```python
from hmcollab import directories
from hmcollab import datasets
from hmcollab import splitter
```

```python
tree = datasets.HMDatasetDirectoryTree()
dataset = datasets.HMDataset(tree=tree, folds='threesets')
```

```python
test_tree = datasets.HMDatasetDirectoryTree(directories.testdata("forty_more_customers"))
test_data = datasets.HMDataset(tree=test_tree)
```

```python
r = np.random.RandomState(42)
```

```python
all_customers = dataset.customers.customer_id.unique()
```

```python
eighty = r.choice(dataset.transactions_y.customer_id.unique(), size=80, replace=False)
```

```python
eighty = set(eighty)
eighty
```

```python
eighty.update(test_data.customers.customer_id.unique())
```

```python
portion = splitter.CustomerPortion(eighty)
pruned_dataset = portion.split(dataset)
```

```python
subdata = datasets.HMDataset(articles=pruned_dataset.articles, customers=pruned_dataset.customers, transactions=pruned_dataset.transactions, folds="threesets")
```

```python
pruned_dataset.customers.to_csv(directories.testdata("larger_dataset/customers.csv"), index=False)
pruned_dataset.articles.to_csv(directories.testdata("larger_dataset/articles.csv"), index=False)
pruned_dataset.transactions.to_csv(directories.testdata("larger_dataset/transactions_train.csv"), index=False)
```
