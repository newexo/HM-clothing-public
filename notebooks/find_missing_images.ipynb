```python
artIDs = set(artIDs)
```

```python
artsWithImgs = [os.path.splitext(os.path.basename(path))[0] for path in glob(directories.data("images/*/*"))]
```

```python
len(artsWithImgs)
```

```python
artsWithoutImgs = artIDs.difference(artsWithImgs)
```

```python
len(artsWithoutImgs)
```

```python
len(artsWithoutImgs.intersection(artIDs))
```
