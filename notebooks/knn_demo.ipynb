```python
from hmcollab import datasets
from hmcollab import articles
```

```python
import numpy as np
import pandas as pd

from IPython.display import Image, display
```

```python
dataset = datasets.HMDataset()
```

```python
simple_munger = articles.ArticleFeaturesSimpleFeatures(dataset.articles)
```

```python
model = articles.ArticleKNN(simple_munger)
```

```python
model.nearest(index=0)
```

```python
_, indices = model.nearest(index=0)
```

```python
indices = indices[0]
```

```python
indices
```

```python
for i in indices[:3]:
    article_id = simple_munger.id_from_index(i)
    filename = dataset.tree.image(article_id)
    display(Image(filename))
```

```python
model.nearest(index=100)
```

```python
_, indices = model.nearest(index=100)
```

```python
indices = indices[0]
```

```python
for i in indices[:3]:
    article_id = simple_munger.id_from_index(i)
    filename = dataset.tree.image(article_id)
    display(Image(filename))
```

```python
for i in indices[-3:]:
    article_id = simple_munger.id_from_index(i)
    filename = dataset.tree.image(article_id)
    display(Image(filename))
```
