```python
from hmcollab import datasets
from hmcollab import articles
from hmcollab import directories
```

```python
dataset = datasets.HMDataset()
```

```python
article_ids = set(dataset.articles.article_id.unique())
transaction_article_ids = set(dataset.transactions.article_id.unique())
```

```python
len(article_ids)
```

```python
len(transaction_article_ids)
```

```python
no_transaction_articles = article_ids.difference(transaction_article_ids)
no_transaction_articles = list(no_transaction_articles)
no_transaction_articles.sort()
```

```python
no_transaction_articles
```

```python
with open(directories.data("no_transaction_articles.txt"), "w") as f:
    for article_id in no_transaction_articles:
        f.write(article_id+"\n")
```

```python
customer_ids = set(dataset.customers.customer_id.unique())
transaction_customer_ids = set(dataset.transactions.customer_id.unique())
```

```python
len(customer_ids)
```

```python
len(transaction_customer_ids)
```

```python
no_transaction_customers = customer_ids.difference(transaction_customer_ids)
no_transaction_customers = list(no_transaction_customers)
no_transaction_customers.sort()
```

```python
with open(directories.data("no_transaction_customers.txt"), "w") as f:
    for customer_id in no_transaction_customers:
        f.write(customer_id+"\n")
```
